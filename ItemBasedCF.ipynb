# Item-based Collaborative Filtering

We will start by loading up the MovieLens dataset. We load the ratings data and the movie data and then merge them so that we can see both the movie name and the ratings from each user.

```python
import pandas as pd
import numpy as np

r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('u.data', sep='\t', names=r_cols, usecols=range(3))

m_cols = ['movie_id', 'title']
movies = pd.read_csv('u.item', sep='|', names=m_cols, usecols=range(2), encoding='latin-1')

ratings = pd.merge(movies, ratings)

ratings.head()
```

We use the pivot table function to create a matrix of users and the movies they rated. NaN values are shown where users did not rate a particular movie.

```python
userRatings = ratings.pivot_table(index=['user_id'], columns=['title'], values = 'rating')
userRatings.head()
```

Next we can use the corr() method to compute a correlation for every movie pair (where at least one user has rated both movies).

```python
corrMatrix = userRatings.corr()
corrMatrix.head()
```

In order to avoid skewed results caused by movies that were only rated by a handful of people, we can use the min_periods argument to only get results from movies where at least 100 people have rated it.

```python
corrMatrix = userRatings.corr(method='pearson', min_periods=100)
corrMatrix.head()
```

Now let's produce some movie recommendations for user ID 0, who has been manually added to the dataset as test case. We will extract his ratings data and drop the missing values.

```python
myRatings = userRatings.loc[0].dropna()
myRatings
```

a

```python
simCandidates = pd.Series()
for i in range(0, len(myRatings.index)):
    print("Adding sims for " + myRatings.index[i] + "...")
    # Retrieve movies similar to this one the user has rated
    sims = corrMatrix[myRatings.index[i]].dropna()
    # Scale the similarity by how well the user rated this movie
    sims = sims.map(lambda x: x * myRatings[i])
    # Add the score to the list of similarity candidates
    simCandidates = simCandidates.append(sims)

# Check results so far:
print("sorting...")
simCandidates.sort_values(inplace=True, ascending=False)
print(simCandidates.head(10))
```

Some movies came up twice because they were similar to more than one movie the user rated. We can use groupby() to add scores together for movies that show up more than once so they will count more.

```python
simCandidates = simCandidates.groupby(simCandidates.index).sum()
```

```python
simCandidates.sort_values(inplace =True, ascending=False)
simCandidates.head(10)
```

After this we filter out the movies that have already been raided by the user.

```python
filteredSims = simCandidates.drop(myRatings.index)
filteredSims.head(10)
```
